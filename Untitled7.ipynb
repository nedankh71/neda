```python
numeric_df =df.select_dtypes(include=['number'])
corrleation_matrix = numeric_df.corr()
corrleation_matrix
```

```python

sns.heatmap(corrleation_matrix, annot=True)
plt.title("Correlation Matrix")
plt.rcParams['figure.figsize'] =(15,10)
plt.show()
```
